```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# convert Kd to free energy
def convert_Kd_to_FE(Kd, kT=2.494):
    """
    Convert Kd to free energy
    Kd: dissociation constant (nM)
    """
    Kd = Kd * 1e-9 # M
    dG = np.log(Kd) * kT
    print(f'Kd = {Kd:.2e} M')
    print(f'dG = {dG:.2f} kJ/mol')
    dG_kcal = dG / 4.184
    print(f'dG = {dG_kcal:.2f} kcal/mol')
```

```python
def convert_FE_to_Kd(FE, kT=2.494):
    """
    Convert free energy to Kd
    FE: free energy (kcal/mol)
    """
    FE = FE * 4.184 # kJ/mol
    Kd = np.exp(-FE / kT)
    print(f'dG = {FE:.2f} kJ/mol')
    print(f'Kd = {Kd :.2e} nM')
```

```python
convert_FE_to_Kd(-22)
```

```python
Kd = np.logspace(-1, 3, 50) # nM
Kd = Kd * 1e-9 # M
dG = np.log(Kd) * kT
dG_kcal = dG / 4.184

fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(Kd*1e9, dG, color='C0')
ax.set_xlabel('Kd (nM)')
ax.set_ylabel('dG (kJ/mol)')
ax.set_xscale('log')
ax.set_title('Kd to dG')

ax_y2 = ax.twinx()
ax_y2.plot(Kd*1e9, dG_kcal, color='C1')
ax_y2.set_ylabel('dG (kcal/mol)')
ax_y2.set_xscale('log')
plt.tight_layout()
plt.show()
```
